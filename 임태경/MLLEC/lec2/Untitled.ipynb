Linear Regression모델의 첫번째 단계
H(x) = Wx + b
H(x) 생각한 가설

가장 비슷한 가설을 찾을때 실제데이터와 생각한 가설의 선의 거리가 낮은 걸 찾음
-> cost function(=loss funcion)
(H(x)-y)제곱 >> 차이가 마이너스인 경우 상관없이 일정하게 차이를 양수로 표현해주고
차이가 클경우 제곱이기때문에 그값을 극대화시켜(패널티를줌) 값을 찾을수있다.

m = 데이터의 개수

Linear Regression모델에서 cost function을 구할때 가장작은 값을 구하는것이 학습이다.
                                                    (=minimize cost(W,b))

# Tensorflow로 간단한 linear regression을 구현


```python
import tensorflow as tf
x_train = [1,2,3]
y_train = [1,2,3]
```

```python
w = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

Variable >> tensorflow가 자체적으로 변경시키는 값

[1] < 값이 하나인 1차원 array(shape설정)

```python
hypothesis = X * w + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

reduce_mean >> 값이 여러개있을때 그것을 평균내주는것

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

for step in range(2001):
    sess.run(train)
    if step % 20 ==0:
        print(step, sess.run(cost), sess.run(w), sess.run(b))

train - cost - hypothesis - w,b

train을 실행시킨다는의미 >> w,b를 값을 조정할수있게됨

```python
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])
```

```python
for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost,w,b,train], feed_dict={X:[1,2,3,4,5], Y:[2.1,3.1,4.1,5.1,6.1]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```
